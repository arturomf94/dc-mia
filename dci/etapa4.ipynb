```python
import pandas as pd
import numpy as np
import sys
import glob
import os.path
import re
import matplotlib.pyplot as plt
```

```python
fig_size = plt.rcParams["figure.figsize"]
plt.rcParams["figure.figsize"] = [15,8]
```

```python
productos = pd.read_pickle('productos_limpia')
productos.shape
```

```python
# Para cada año, identificar las cantidades máximas y mínimas de los productos que se distribuyen.

# MAX
print('2012')
productos_name = []
for producto in productos.loc[productos['ANIO'] == '2012']['PRODUCTO']:
    if not producto in productos_name:
        productos_name.append(producto)
filter_aux = productos.loc[productos['ANIO'] == '2012']
productos_max = []
for producto in productos_name:
    prod_max = pd.to_numeric(filter_aux.loc[filter_aux['PRODUCTO'] == producto]['VENTA TOTAL'], errors = 'coerce').max()
    productos_max.append(prod_max)
plt.bar(productos_name, productos_max)
plt.xticks(rotation=90)
plt.title('Max Venta Total por Producto en 2012')
plt.ylabel('Cantidad')
plt.show()

# MIN
print('2012')
productos_name = []
for producto in productos.loc[productos['ANIO'] == '2012']['PRODUCTO']:
    if not producto in productos_name:
        productos_name.append(producto)
filter_aux = productos.loc[productos['ANIO'] == '2012']
productos_min = []
for producto in productos_name:
    prod_min = pd.to_numeric(filter_aux.loc[filter_aux['PRODUCTO'] == producto]['VENTA TOTAL'], errors = 'coerce').min()
    productos_min.append(prod_min)
plt.bar(productos_name, productos_min)
plt.xticks(rotation=90)
plt.title('Min Venta Total por Producto en 2012')
plt.ylabel('Cantidad')
plt.show()
```

```python
folder_name = "./data_4/"

data_path = os.path.join(folder_name,'*csv')

files = glob.glob(data_path)
files
```

```python
file = files[0]
materia_prima = pd.read_csv(file)
```

```python
materia_prima.head()
```

```python
# Cambiamos NaNs por 0s 
materia_prima = materia_prima.fillna(0)
materia_prima.head()
```

```python
# Eliminamos encabezados
materia_prima=materia_prima[materia_prima.DIA != 'DIA']
materia_prima.head()
```

```python
# Agregamos una columna para indicar la semana 
materia_prima['SEMANA'] = 1

inc = 0
contador = 0
for i_f, f in  materia_prima.iterrows():
    if(contador == 7):
        contador = 0
        inc += 1
    materia_prima.loc[i_f, 'SEMANA'] += inc
    contador += 1
materia_prima
```
