```python
import pandas as pd
import sys
import glob
import os.path
```

```python
folder_name = "./data/"

data_path = os.path.join(folder_name,'*xlsx')

files = glob.glob(data_path)
```

```python
files
```

```python
data_frames = []
for file in files:
    df = pd.read_excel(file, sheet_name = None)
    data_frames.append(df)
```

```python
test = data_frames[0]['Hoja1']
test
```

```python
cols = ['PRODUCTO', 'CARGA', 'EXISTENCIA', 'DEVOLUCION', 'VENTA TOTAL', 'PRECIO * PRODUC', '$', 'VENTA EN LT', 'RUTA', 'DIA', 'SEMANA']
productos = pd.DataFrame(columns= cols)
```

```python
rutas = []
semanas = []
dias = []
test_copy = test
for i in range(int(test.shape[0])):
    for j in range(int(test.shape[1])):
        if type(test.iloc[i, j]) == unicode and 'RUTA:' in test.iloc[i, j]:
            ruta = ''
            dia = ''
            semana = ''
            ruta = test.iloc[i, j]
            if not(test.iloc[i, j] in rutas):
                rutas.append(ruta)
            if type(test.iloc[i - 1,j]) == unicode:
                dia = test.iloc[i - 1, j]
                if not(test.iloc[i - 1, j] in dias):
                    dias.append(dia)
            if type(test.iloc[i - 2,j]) == unicode and 'SEMANA' in test.iloc[i - 2,j]:
                semana = test.iloc[i - 2, j]
                if not(test.iloc[i - 2, j] in semanas):
                    semanas.append(semana)
            new_entry = test_copy.iloc[i + 2:i + 10, j:j + 1].T
            new_entry['RUTA'] = ruta
            new_entry['DIA'] = dia
            new_entry['SEMANA'] = semana
            new_entry.columns = cols
            # print(new_entry)            
            # print(list(new_entry.columns.values))
            if ruta != '' and not 'QUESO' in ruta:
                productos = productos.append(new_entry, ignore_index=True)
```
